# ***Data Preprocessing***


```python
df.drop(columns=["Native Country", "Education"], inplace=True)
```

```python
df.info()
```

```python
 df["Final Weight"].value_counts()
```

```python
df.drop(columns=["Final Weight"], inplace=True)
```

```python
label_encoder = LabelEncoder()
df["Gender"] = label_encoder.fit_transform(df["Gender"])
df["Income"] = label_encoder.fit_transform(df["Income"])
```

```python
def ohe(data,cols, drop_first=True):
    encoded_data = df.copy()

    for col in cols:
        dumm = pd.get_dummies(data[col], dtype=int, drop_first=drop_first)
        del encoded_data[col]
        encoded_data = pd.concat([encoded_data, dumm], axis=1)

    return encoded_data

df = ohe(df, ['Workclass', 'Marital Status', 'Occupation', 'Relationship', 'Race'])
```

```python
df.head()
```

```python
df["Income"].value_counts()
```

#***Splitting the Data***

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=["Income"]), df["Income"], test_size=0.2, random_state=42)
```

#Classification Using Machine Learning Models

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    #"Support Vector Machine": SVC(),
    "XGBoost": xgb.XGBClassifier()
}
```

```python
for model_name, model in models.items():
  model.fit(X_train, y_train)
  print(f"{model_name} Classification Report:")
  print(classification_report(y_test, model.predict(X_test)))
  print("___" * 50 + "\n\n")
```

#Classification Using Deep Neural Network

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential(
    [
     Dense(32, activation="relu"),
     Dense(128, activation="relu"),
     Dense(364, activation="relu"),
     Dense(1, activation="sigmoid")
    ]
)

early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    verbose=1,
    restore_best_weights=True)

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
model.fit(X_train, y_train, epochs=200, batch_size=32, validation_data=(X_test, y_test) , callbacks=[early_stopping])
```
